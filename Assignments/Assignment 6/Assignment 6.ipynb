```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data=pd.read_csv("Data.csv")
Data.head()
```

```python
# Looking at the Data
print(Data.columns.tolist())
Data.describe(include="all")
```

```python
sns.scatterplot(data=Data,x="MaxTemp",y="Rainfall")
plt.show()
```

```python
# Visualisation 2
# Seeing if Sunshine today affects the chances of rain tommorow

sns.boxplot(data=Data,y="Sunshine",x="RainTomorrow")
plt.show()
```

```python
#Visualisation 3
# Seeing if there is a relation between Humidity at 3pm and 9am

sns.jointplot(data=Data,kind="hex",x='Humidity3pm',y="Humidity9am")
plt.show()
```

```python
#Visualisation 4
# Getting the Distribution of Rainfall over Data
sns.distplot(Data['Rainfall'])
plt.show()
```

```python
import re
exp="^(.+?)/"
for i in range(0,Data.shape[0]):
    res=re.findall(exp,Data['Date'][i])
    Data["Month"][i]=res[0]
```

```python
# Visualisation 18
# Let's See how the Month of year is related with Rainfall
sns.swarmplot(data=Data,y="Month",x="RainToday")
plt.show()
```

```python
# Visualisation 19 and 20
# Let's See if The months affect min and max temp!
sns.scatterplot(data=Data,x="MinTemp",y="Month",hue="RainToday")
plt.show()
sns.scatterplot(data=Data,x="MaxTemp",y="Month",hue="RainToday")
plt.show()
```
